# Clustering

### Problem Statement: In this assignment I'm going to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference.

```python
#Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#from sklearn import cluster, datasets
from sklearn.cluster import KMeans
from scipy import misc
import scipy.misc
```

```python
#Load the dataset from scipy.misc and Visualizing the raw image before compressing
Racoon = scipy.misc.face(gray=True)
plt.figure(figsize=(8,8))
plt.imshow(Racoon,plt.cm.gray)
plt.show()
```

```python
print(f"Rows: {Racoon.shape[0]}, Columns: {Racoon.shape[1]}")
```

```python
rows = Racoon.shape[0]
columns = Racoon.shape[1]
image  = Racoon.reshape(rows*columns,1)
```

### Using KMeans to fit the image

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(image)
```

Getting the cluster and the labels for the image

```python
clusters = kmeans.cluster_centers_
labels = kmeans.labels_
labels = labels.reshape(rows,columns)
```

```python
print(clusters)
```

```python
#Save the compressed image
plt.imsave("Racoon_compressed.png",labels)
```

```python
#Visualize the compresssed image
compressed_image = plt.imread('Racoon_compressed.png')
plt.figure(figsize=(8,8))
plt.imshow(compressed_image)
plt.show()
```
